```r
likelihoord_function= function(E1,P1,P2,P3,P4,P5){
    Q=matrix(0,5,5)
    Q[1,1]=P1
    Q[2,2]=P2
    Q[3,3]=P3
    Q[4,4]=P4
    Q[5,5]=P5
    H=E1
    R=1
    T=1
    df=data.frame(year=as.Date(as.yearqtr(time(m1gr))),m1gr,dint_lag,inf_lag,surp_lag,m1gr_lag)
    
    y=c(0,length(m1gr))
    for (i in 1:length(m1gr)){
        y[i]=df[,2][i]
    }
    Z=array(1,c(row=1,col=5,length(m1gr)))
    for (i in 1:length(m1gr)){
        Z[1,2,i]=df[,3][i]
        Z[1,3,i]=df[,4][i]
        Z[1,4,i]=df[,5][i]
        Z[1,5,i]=df[,6][i]
    }

    a=array(0,c(5,1,length(m1gr)))
    P=array(0,c(5,5,length(m1gr)))
    v=c(0,length(m1gr))
    F=c(0,length(m1gr))
    K=array(0,c(5,1,length(m1gr)))
    lik=0
    for (i in length(m1gr)){

        v[i]= y[i] - t(Z[,,i]) %*% a[,,i]
        F[i]= t(Z[,,i]) %*% P[,,i] %*% t(t(Z[,,i])) + H
        K[,,i]= T %*% P[,,i] %*% t(t(Z[,,i]))%*% inv(F[i])
        a[,,i+1]= T %*% a[,,i] + K[,,i] %*% v[i]
        P[,,i+1]= T %*% P[,,i] %*% t(T-K[,,i] %*% t(Z[,,i])) + R %*% Q %*% t(R) 
        lik= lik + log(det(F[i]))+v[i]%*%inv(F[i])%*%v[i]
    }
return(lik)}
```

```r
likelihood_function= function(E1,P1,P2,P3,P4,P5){
  Q=matrix(0,5,5)
  Q[1,1]=P1
  Q[2,2]=P2
  Q[3,3]=P3
  Q[4,4]=P4
  Q[5,5]=P5
  H=E1
  R=1
  Tt=1
  df=data.frame(year=as.Date(as.yearqtr(time(m1gr))),m1gr,dint_lag,inf_lag,surp_lag,m1gr_lag)
  
  y=c(0,length(m1gr))
  for (i in 1:length(m1gr)){
    y[i]=df[,2][i]
  }
  Z=array(1,c(row=1,col=5,length(m1gr)))
  for (i in 1:length(m1gr)){
    Z[1,2,i]=df[,3][i]
    Z[1,3,i]=df[,4][i]
    Z[1,4,i]=df[,5][i]
    Z[1,5,i]=df[,6][i]
  }
    
  a=array(0,c(5,1,length(m1gr)+1))
  P=array(0,c(5,5,length(m1gr)+1))
  v=NULL
  Ft=NULL  
  K=array(0,c(5,1,length(m1gr)))
  lik=0
  for (i in 1:length(m1gr)){
      
    v[i]= y[i] - t(Z[,,i]) %*% a[,,i]
    Ft[i]= t(Z[,,i]) %*% P[,,i] %*% t(t(Z[,,i])) + H
    K[,,i]= Tt * P[,,i] %*% t(t(Z[,,i])) * (Ft[i])^(-1)
    a[,,i+1]= Tt * a[,,i] + K[,,i] * v[i]
    P[,,i+1]= Tt * P[,,i] %*% t(Tt - K[,,i] %*% t(Z[,,i])) + R * Q * (R) 
    lik= lik + log(abs(Ft[i]))+ v[i] * (Ft[i])^(-1) * v[i]
  }
    return(lik)
    
    
  }
```
