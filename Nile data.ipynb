```r
lik_func= function(E,N){
    y=NULL
    for(i in 1:length(Nile)){
        y[i]=Nile[i]
    }
    a=NULL
    a[1]=0
    P=NULL
    P[1]=10^7
    v=NULL
    lik=0
    K=NULL
    Ft=NULL
    for(i in 1:length(Nile)){
        v[i]= y[i]- a[i]
        Ft[i]= P[i] + E
        K[i]= P[i]/Ft[i]
        a[i+1]= a[i]+ K[i]*v[i]
        P[i+1]= P[i]*(1-K[i])+ N
        lik= lik+log(Ft[i])+((v[i])^2)/Ft[i]
    }
    my_list= list(v=v,Ft=Ft,a=a,P=P,likelihood=(-(((length(Nile))/2)*log(2*pi))-(1/2)*lik))
    return(my_list)
}
```
