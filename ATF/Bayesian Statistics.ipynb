>1. The **“prior,” p(θ)**, is the credibility of the θ values without the data D. 
>2. The **“posterior,” p(θ|D)**, is the credibility of θ values with the data D taken into account. 
>3. The **“likelihood,” p(D|θ)**, is the probability that the data could be generated by the model with parameter value θ. 
>4. The **“evidence”** for the model, **p(D)**, is the overall probability of the data according to the model, determined by averaging across all possible parameter values weighted by the strength of belief in those parameter values. Also known as **marginal likelihood**
