```r
library(ggplot2)
set.seed(123456)
sims=1000
alpha.1=numeric(sims)
B.1=numeric(sims)
n=100
a=5
b=seq(1.1,2,0.1) # a vector of different betas
X=runif(n,-1,1) # x follows uniform distribution

reject=rep(0,10) # vector of number of rejections of the hypothesis that beta is 1
accept=rep(0,10) # vector of number of acceptences of the hypothesis that beta is 1

for (j in 1:10){
    for(i in 1:sims){
    Y=a+b[j]*X+rnorm(n,0,1) # DGP
    model=lm(Y~X) # linear reg
    B.1=model$coef[2] # sample beta
    v=vcov(model) # Variance covariance matrix 
    se=sqrt(v[2,2]) # standard error of sample beta
    if (((B.1-1)/se) > 1.96){ # t-stat and t-test
        reject[j]=reject[j]+1
    
    }
    else{
        accept[j]=accept[j]+1
    }
    
}
}
accept=accept/1000
reject=reject/1000
df=data.frame(accept,reject,b)
df
# ploting acceptance and rejection
plot=ggplot(df,aes(x=b))
plot.1=plot+ geom_point(aes(y=accept,col='red')) + scale_color_discrete(name='',labels=c('Reject','Accept'))
plot.2=plot.1+geom_point(aes(y=reject,col='blue'))
plot.2+labs(title='Power Test', x='Beta',y='Ratio')
```
